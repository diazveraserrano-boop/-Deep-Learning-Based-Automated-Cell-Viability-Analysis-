```python
# Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

import os
import re

# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
# CONFIGURACI√УN - CAMBIA ESTAS RUTAS SI ES NECESARIO
# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
CARPETA_IMAGENES_CRUDAS    = '/content/drive/MyDrive/FINAL_IMAGES(sin_augmentation)'
CARPETA_MASCARAS           = '/content/drive/MyDrive/FINAL_MASKS(sin_augmentation)'

# N√Їmero de d√≠gitos que quieres para el relleno con ceros
# Ejemplos:
#   3 вЖТ Image001.jpg, Image010.jpg, Image100.jpg
#   4 вЖТ Image0001.jpg, Image0010.jpg, Image0100.jpg
DIGITOS = 4

# Prefijos y sufijo que est√°s usando
PREFIJO_IMAGEN    = "Image"
PREFIJO_MASCARA   = "Image"
SUFIJO_MASCARA    = "mask"

# Extensiones que consideramos
EXTENSIONES = ('.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG')

# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
# Funci√≥n para extraer el n√Їmero del nombre
# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
def extraer_numero(nombre):
    coincidencia = re.search(r'\d+', nombre)
    if coincidencia:
        return int(coincidencia.group())
    return None

# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
# 1. Procesar carpeta de im√°genes crudas
# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
print("Procesando im√°genes crudas...\n")

archivos_crudos = [f for f in os.listdir(CARPETA_IMAGENES_CRUDAS)
                   if os.path.isfile(os.path.join(CARPETA_IMAGENES_CRUDAS, f))
                   and f.lower().endswith(EXTENSIONES)]

# Ordenamos por n√Їmero extra√≠do
archivos_crudos_ordenados = sorted(archivos_crudos, key=lambda x: extraer_numero(x) or 0)

for nombre_viejo in archivos_crudos_ordenados:
    numero = extraer_numero(nombre_viejo)
    if numero is None:
        print(f"  Ignorado (sin n√Їmero): {nombre_viejo}")
        continue

    extension = os.path.splitext(nombre_viejo)[1]
    nombre_nuevo = f"{PREFIJO_IMAGEN}{numero:0{DIGITOS}d}{extension}"

    ruta_vieja = os.path.join(CARPETA_IMAGENES_CRUDAS, nombre_viejo)
    ruta_nueva = os.path.join(CARPETA_IMAGENES_CRUDAS, nombre_nuevo)

    if nombre_viejo != nombre_nuevo:
        if os.path.exists(ruta_nueva):
            print(f"  ¬°Conflicto! Ya existe: {nombre_nuevo} вЖТ saltando {nombre_viejo}")
        else:
            os.rename(ruta_vieja, ruta_nueva)
            print(f"  {nombre_viejo} вЖТ {nombre_nuevo}")
    else:
        print(f"  Ya est√° correcto: {nombre_viejo}")

# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
# 2. Procesar carpeta de m√°scaras
# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
print("\nProcesando m√°scaras...\n")

archivos_mascaras = [f for f in os.listdir(CARPETA_MASCARAS)
                     if os.path.isfile(os.path.join(CARPETA_MASCARAS, f))
                     and f.lower().endswith(EXTENSIONES)]

# Ordenamos por n√Їmero extra√≠do
archivos_mascaras_ordenados = sorted(archivos_mascaras, key=lambda x: extraer_numero(x) or 0)

for nombre_viejo in archivos_mascaras_ordenados:
    numero = extraer_numero(nombre_viejo)
    if numero is None:
        print(f"  Ignorado (sin n√Їmero): {nombre_viejo}")
        continue

    extension = os.path.splitext(nombre_viejo)[1]
    nombre_nuevo = f"{PREFIJO_MASCARA}{numero:0{DIGITOS}d}{SUFIJO_MASCARA}{extension}"

    ruta_vieja = os.path.join(CARPETA_MASCARAS, nombre_viejo)
    ruta_nueva = os.path.join(CARPETA_MASCARAS, nombre_nuevo)

    if nombre_viejo != nombre_nuevo:
        if os.path.exists(ruta_nueva):
            print(f"  ¬°Conflicto! Ya existe: {nombre_nuevo} вЖТ saltando {nombre_viejo}")
        else:
            os.rename(ruta_vieja, ruta_nueva)
            print(f"  {nombre_viejo} вЖТ {nombre_nuevo}")
    else:
        print(f"  Ya est√° correcto: {nombre_viejo}")

print("\n" + "="*60)
print("Renombrado con padding finalizado.")
print(f"Ahora los archivos deber√≠an verse ordenados como:")
print(f"  Image001.jpg")
print(f"  Image002.jpg")
print(f"  ...")
print(f"  Image010.jpg")
print(f"  Image011.jpg")
print("="*60)

print("\nRefresca la vista de archivos en Drive o en el panel lateral de Colab para ver el orden correcto.")
```

```python
from PIL import Image
import os
import re

# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
# CONFIGURACI√УN
# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
CARPETA_CRUDAS     = '/content/drive/MyDrive/FINAL_IMAGES(sin_augmentation)'
CARPETA_MASCARAS   = '/content/drive/MyDrive/FINAL_MASKS(sin_augmentation)'

DIGITOS            = 4
PREFIJO            = "Image"
SUFIJO_MASCARA     = "mask"
EXTENSION_CRUDA    = ".jpg"
EXTENSION_MASCARA  = ".jpg"  # originales .jpg, nuevas m√°scaras en .png

# вФАвФА Transformaciones visibles вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
SHIFTS_PCT = [(0.25, 0), (0, 0.25)]  # +25% x y +25% y
FLIPS = ['horizontal', 'vertical']
ROTACIONES = [i * 18 for i in range(20)]  # 0¬∞, 18¬∞, ..., 342¬∞
RECORTES = [0.70, 0.80, 0.90]  # 70%, 80%, 90% crop + resize

# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
# Funciones auxiliares
# вФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФАвФА
def extraer_numero(nombre):
    m = re.search(r'\d+', nombre)
    return int(m.group()) if m else 0

def formato_con_padding(numero, digitos):
    return f"{numero:0{digitos}d}"

# Encontrar m√°ximo original
nums_originales = []
for f in os.listdir(CARPETA_CRUDAS):
    if f.startswith(PREFIJO) and f.endswith(EXTENSION_CRUDA):
        num = extraer_numero(f)
        if num > 0:
            nums_originales.append(num)

if not nums_originales:
    print("No se encontraron im√°genes originales.")
else:
    max_original = max(nums_originales)
    print(f"Se detectaron {len(nums_originales)} im√°genes originales (hasta Image{formato_con_padding(max_original, DIGITOS)})\n")

    siguiente_num = max_original + 1
    print(f"Las nuevas empezar√°n desde: Image{formato_con_padding(siguiente_num, DIGITOS)}\n")

    contador_nuevas = 0

    for i in range(1, max_original + 1):
        num_str = formato_con_padding(i, DIGITOS)
        cruda_nombre = f"{PREFIJO}{num_str}{EXTENSION_CRUDA}"
        ruta_cruda = os.path.join(CARPETA_CRUDAS, cruda_nombre)

        if not os.path.exists(ruta_cruda):
            continue

        mascara_nombre = f"{PREFIJO}{num_str}{SUFIJO_MASCARA}{EXTENSION_MASCARA}"
        ruta_mascara = os.path.join(CARPETA_MASCARAS, mascara_nombre)

        if not os.path.exists(ruta_mascara):
            print(f"Falta m√°scara para {cruda_nombre} вЖТ saltando")
            continue

        try:
            img  = Image.open(ruta_cruda).convert("RGB")
            mask = Image.open(ruta_mascara).convert("RGB")
        except Exception as e:
            print(f"Error cargando {cruda_nombre}: {e}")
            continue

        print(f"Procesando original: {cruda_nombre}")

        w, h = img.size

        # 1. 2 Shifts (traslaciones grandes)
        for dx_pct, dy_pct in SHIFTS_PCT:
            dx = int(w * dx_pct)
            dy = int(h * dy_pct)

            img_shift = Image.new("RGB", (w, h), (0, 0, 0))
            mask_shift = Image.new("RGB", (w, h), (0, 0, 0))

            img_shift.paste(img, (dx, dy))
            mask_shift.paste(mask, (dx, dy))

            nuevo_str = formato_con_padding(siguiente_num, DIGITOS)
            nuevo_cruda = f"{PREFIJO}{nuevo_str}{EXTENSION_CRUDA}"
            nuevo_mask = f"{PREFIJO}{nuevo_str}{SUFIJO_MASCARA}.png"

            img_shift.save(os.path.join(CARPETA_CRUDAS, nuevo_cruda))
            mask_shift.save(os.path.join(CARPETA_MASCARAS, nuevo_mask), format='PNG')

            print(f"   вФФвФА shift {dx_pct*100:.0f}%x {dy_pct*100:.0f}%y вЖТ {nuevo_cruda}")
            siguiente_num += 1
            contador_nuevas += 1

        # 2. 2 Flips
        for flip_type in FLIPS:
            if flip_type == 'horizontal':
                img_f = img.transpose(Image.FLIP_LEFT_RIGHT)
                mask_f = mask.transpose(Image.FLIP_LEFT_RIGHT)
            else:  # vertical
                img_f = img.transpose(Image.FLIP_TOP_BOTTOM)
                mask_f = mask.transpose(Image.FLIP_TOP_BOTTOM)

            nuevo_str = formato_con_padding(siguiente_num, DIGITOS)
            nuevo_cruda = f"{PREFIJO}{nuevo_str}{EXTENSION_CRUDA}"
            nuevo_mask = f"{PREFIJO}{nuevo_str}{SUFIJO_MASCARA}.png"

            img_f.save(os.path.join(CARPETA_CRUDAS, nuevo_cruda))
            mask_f.save(os.path.join(CARPETA_MASCARAS, nuevo_mask), format='PNG')

            print(f"   вФФвФА flip {flip_type} вЖТ {nuevo_cruda}")
            siguiente_num += 1
            contador_nuevas += 1

        # 3. 20 Rotaciones
        for ang in ROTACIONES:
            img_rot = img.rotate(ang, resample=Image.BICUBIC, expand=False)
            mask_rot = mask.rotate(ang, resample=Image.NEAREST, expand=False)

            nuevo_str = formato_con_padding(siguiente_num, DIGITOS)
            nuevo_cruda = f"{PREFIJO}{nuevo_str}{EXTENSION_CRUDA}"
            nuevo_mask = f"{PREFIJO}{nuevo_str}{SUFIJO_MASCARA}.png"

            img_rot.save(os.path.join(CARPETA_CRUDAS, nuevo_cruda))
            mask_rot.save(os.path.join(CARPETA_MASCARAS, nuevo_mask), format='PNG')

            print(f"   вФФвФА rot {ang:+3}¬∞  вЖТ {nuevo_cruda}")
            siguiente_num += 1
            contador_nuevas += 1

        # 4. 3 Recortes centrados + resize
        for pct in RECORTES:
            crop_w = int(w * pct)
            crop_h = int(h * pct)
            left = (w - crop_w) // 2
            top = (h - crop_h) // 2
            right = left + crop_w
            bottom = top + crop_h

            img_crop = img.crop((left, top, right, bottom))
            mask_crop = mask.crop((left, top, right, bottom))

            img_res = img_crop.resize((w, h), Image.BICUBIC)
            mask_res = mask_crop.resize((w, h), Image.NEAREST)

            nuevo_str = formato_con_padding(siguiente_num, DIGITOS)
            nuevo_cruda = f"{PREFIJO}{nuevo_str}{EXTENSION_CRUDA}"
            nuevo_mask = f"{PREFIJO}{nuevo_str}{SUFIJO_MASCARA}.png"

            img_res.save(os.path.join(CARPETA_CRUDAS, nuevo_cruda))
            mask_res.save(os.path.join(CARPETA_MASCARAS, nuevo_mask), format='PNG')

            print(f"   вФФвФА crop {pct*100:.0f}% вЖТ {nuevo_cruda}")
            siguiente_num += 1
            contador_nuevas += 1

    print("\n" + "вХР"*70)
    print("AUMENTACI√УN FINALIZADA")
    print(f"Variantes por original: 27 (2 shifts + 2 flips + 20 rot + 3 crops)")
    print(f"Nuevas im√°genes totales: {contador_nuevas}")
    print(f"Dataset aproximado ahora: {len(nums_originales) + contador_nuevas}")
    print(f"√Ъltima imagen creada: Image{formato_con_padding(siguiente_num-1, DIGITOS)}")
    print("Todas las m√°scaras nuevas en .png")
    print("вХР"*70)
    print("Refresca las carpetas en Drive. Descarga algunas nuevas y comp√°ralas.")
```
